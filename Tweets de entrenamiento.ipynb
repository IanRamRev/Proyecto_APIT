### Se utilizaron tweets ya etiquetados para comprobar la clasificación de nuestro dataset de tweets

```python
import numpy
import pandas
from bs4 import BeautifulSoup
import csv
```

```python
with open('TASS2019_country_MX_train.xml', 'rb') as html:
    soup = BeautifulSoup(html)
```

```python
tweets = []
for tweet in soup.find_all('content'):
    aux = tweet.get_text()
    tweets.append(aux)
    
polaridad = []
for value in soup.find_all('value'):
    aux = value.get_text()
    polaridad.append(aux)
    
p = numpy.asarray(polaridad, dtype=str)
print(numpy.unique(p))
print("Negativo = ", numpy.sum(p == p[0]))
print("Neutral = ", numpy.sum(p == p[1]))
print("NONE = ", numpy.sum(p == p[2]))
print("Positivo = ", numpy.sum(p == p[3]))
```

```python
data = []
for i in range(len(tweets)):
    data.append([tweets[i], polaridad[i]])
    
data
```

```python
myFile = open('TweetsClasificados.csv', 'w', encoding='utf-8')
with myFile:
    writer = csv.writer(myFile)
    writer.writerows(data)
```
